# Batch Gradient Decent 

```python
import time
start = time.time()

eta= 0.1             #learning rate
n_iterations = 1000
m=100                             #no of training instances

theta= np.random.randn(2,1)   #"standard normal" distribution value of theta (mean=0,var=1)
print(f"random initialized value of theta : {theta}")

for iteration in range(n_iterations):
    print(f"iteration number : {iteration}")
    gradiants = 2/m * np.dot(X_with_bias.T , (np.dot(X_with_bias, theta) - y))
    print(f"the value of gradiants: {gradiants}")
    theta = theta - eta*gradiants
    print(f"value of theta in {iteration} no is :{theta}")
    
end = time.time()-start
print(f"total time {end}")
```

# Min Batch Gradiant Decent

```python
start = time.time()
theta_path_mgd = []

n_iterations = 50
minibatch_size = 20

np.random.seed(42)
theta = np.random.randn(2,1)  # random initialization
print(f"theta {theta}")

t0, t1 = 200, 1000
def learning_schedule(t):
    return t0 / (t + t1)

m = 100
t = 0
for epoch in range(n_iterations):
    shuffled_indices = np.random.permutation(m)
    
    X_b_shuffled = X_new[shuffled_indices]
    y_shuffled = y[shuffled_indices]
    
    for i in range(0, m, minibatch_size):
        print(f"batch no i {i}")
        t += 1
        xi = X_b_shuffled[i:i+minibatch_size]
        yi = y_shuffled[i:i+minibatch_size]
        print(f"shape {xi.shape}")
        
        gradients = 2/minibatch_size * xi.T.dot(xi.dot(theta) - yi)
        print(f"gradiant {gradients}")
        
        eta = learning_schedule(t)
        print(f"eta {eta}")
        theta = theta - eta * gradients
        print(f"theta {theta}")
        theta_path_mgd.append(theta)

print(f"theta {theta}")
end = time.time()-start
print(f"total time {end}")
```
